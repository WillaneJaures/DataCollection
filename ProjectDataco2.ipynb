```python
#on import  les libraries necessaire pour scraper nos donnees

import numpy as np
import pandas as pd
from requests import get # get nous permettra de recuperer les url de toutes pages
from bs4 import BeautifulSoup as bs  ## beautiful mous permetra de stocker le code html en objet Beautifulsoup
```

```python

url = 'https://sn.coinafrique.com/categorie/appartements'
resp = get(url) # on recupere le code html
soup = bs(resp.text, 'html.parser') ## on stock le code html en objet bs
links_a = soup.find_all('a', class_ = 'card-image ad__card-image waves-block waves-light') # on slection tous les tags <a> sw class = 'card-image ad__card-image waves-block waves-light'
links1 = [link['href'] for link in links_a] #on recupere tous les href pour tous les element dans votre variable links_a
links = ['https://sn.coinafrique.com' + s for s in links1] # les donnees n'ayant pas de liens corrects, on ajoute https://sn.coinafrique.com pour que les liens soient tous correct
links
```

```python
#commencons a scraper les donnees d'un seul lien

link = links[0]
res = get(link)
Soup = bs(res.text, 'html.parser')
inf = Soup.find_all('span', class_ = 'qt')
nombres_de_pieces = inf[0].text
prix = Soup.find('p', class_ = 'price').text.replace('CFA','').replace(' ', '')
addresse1 = Soup.find('p', class_ = 'extras').find_all('span', class_ = 'valign-wrapper')
addresse = addresse1[1].text
image_lien = soup.find('img', class_ = 'ad__card-img')['src']
```

```python
# Scrapons les donnees de tous les liens
data = [] # on cree un tableau vide qui contiendra nos variable
for link in links:
  res = get(link)
  Soup = bs(res.text, 'html.parser')
  try:
    inf = Soup.find_all('span', class_ = 'qt')
    nombres_de_pieces = inf[0].text
    prix = Soup.find('p', class_ = 'price').text.replace('CFA','').replace(' ', '')
    addresse1 = Soup.find('p', class_ = 'extras').find_all('span', class_ = 'valign-wrapper')
    addresse = addresse1[1].text
    image_lien = soup.find('img', class_ = 'ad__card-img')['src']

    #creation du dictionnnaire qui contiendra nos varible
    obj = {
        'nombres_de_pieces' : nombres_de_pieces,
        'prix' : prix,
        'addresse': addresse,
        'image_lien': image_lien
    }

    data.append(obj) # on append les donnees collectees de dans dictionnaire obj dans notre list Data


  except:
    pass
DF = pd.DataFrame(data) #on convertis notre tableau en dataFrame
```

```python
#Scrapons sur toutes les pages
df = pd.DataFrame() #creation du nouvead dataframe qui contiendra les donnees de toutes les pages
for p in range(1,10):
  url = f'https://sn.coinafrique.com/categorie/appartements?page={p}'
  resp = get(url)
  soup = bs(resp.text, 'html.parser')
  links_a = soup.find_all('a', class_ = 'card-image ad__card-image waves-block waves-light')
  links1 = [link['href'] for link in links_a]
  links = ['https://sn.coinafrique.com' + s for s in links1] # les donnees n'ayant pas de link correct en ajoutant le https

  data = []
  for link in links:
    res = get(link)
    Soup = bs(res.text, 'html.parser')
    try: # on fait passer toutes les variable qu'on a cree
      inf = Soup.find_all('span', class_ = 'qt')
      nombres_de_pieces = inf[0].text
      prix = Soup.find('p', class_ = 'price').text.replace('CFA','').replace(' ', '')
      addresse1 = Soup.find('p', class_ = 'extras').find_all('span', class_ = 'valign-wrapper')
      addresse = addresse1[1].text
      image_lien = soup.find('img', class_ = 'ad__card-img')['src']

      #creation du dictionnnaire qui contiendra nos varible
      obj = {
          'nombres_de_pieces' : nombres_de_pieces,
          'prix' : prix,
          'addresse': addresse,
          'image_lien': image_lien
      }

      data.append(obj)


    except:
      pass
  # On joint nos data dataframe DF et df
  DF = pd.DataFrame(data)
  df = pd.concat([DF,df], axis = 0).reset_index(drop = True)
```

  Scraper les donnees des page en utilisant beautiful soup

  url 1 : https://sn.coinafrique.com/categorie/terrains

  Varibable1: superficie, V2 : prix, Varibale3 : adresse , Varibale4 : image lien

```python
import numpy as np
import pandas as pd
from requests import get
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://sn.coinafrique.com/categorie/terrains'
resp = get(url)
soup = bs(resp.text, 'html.parser')
links_a = soup.find_all('a', class_ = 'card-image ad__card-image waves-block waves-light')
links1 = [link['href'] for link in links_a]
links = ['https://sn.coinafrique.com' + s for s in links1] # les donnees n'ayant pas de link correct en ajoutant le https
```

```python
link = links[0]
res = get(link)
Soup = bs(res.text, 'html.parser')
superficie = Soup.find('span', class_ = 'qt').text
prix = Soup.find('p', class_ = 'price').text.replace('CFA','').replace(' ','')
addresse1 = Soup.find('p', class_ = 'extras').find_all('span', class_ = 'valign-wrapper')
addresseo = addresse1[1].text.split(",")
quartier = addresseo[0]
ville = addresseo[1]
image_lien = soup.find('img', class_ = 'ad__card-img')['src']

```

```python
data = []
for link in links:
  res = get(link)
  Soup = bs(res.text, 'html.parser')
  try:
    superficie = Soup.find('span', class_ = 'qt').text
    prix = Soup.find('p', class_ = 'price').text.replace('CFA','').replace(' ','')
    addresse1 = Soup.find('p', class_ = 'extras').find_all('span', class_ = 'valign-wrapper')
    addresseo = addresse1[1].text.split(",")
    quartier = addresseo[0]
    ville = addresseo[1]
    image_lien = soup.find('img', class_ = 'ad__card-img')['src']

    obj = {
        'supercie': superficie,
        'prix': prix,
        'quartier': quartier,
        'ville': ville,
        'image_lien': image_lien
    }

    data.append(obj)

  except:
    pass
DF = pd.DataFrame(data)
```

```python
df = pd.DataFrame()
for p in range(1,10):
  url = f'https://sn.coinafrique.com/categorie/terrains?page={p}'
  resp = get(url)
  soup = bs(resp.text, 'html.parser')
  links_a = soup.find_all('a', class_ = 'card-image ad__card-image waves-block waves-light')
  links1 = [link['href'] for link in links_a]
  links = ['https://sn.coinafrique.com' + s for s in links1] # les donnees n'ayant pas de link correct en ajoutant le https

  data = []
  for link in links:
    res = get(link)
    Soup = bs(res.text, 'html.parser')
    try:
      superficie = Soup.find('span', class_ = 'qt').text
      prix = Soup.find('p', class_ = 'price').text.replace('CFA','').replace(' ','')
      addresse1 = Soup.find('p', class_ = 'extras').find_all('span', class_ = 'valign-wrapper')
      addresseo = addresse1[1].text.split(",")
      quartier = addresseo[0]
      ville = addresseo[1]
      image_lien = soup.find('img', class_ = 'ad__card-img')['src']

      obj = {
          'supercie': superficie,
          'prix': prix,
          'quartier': quartier,
          'ville': ville,
          'image_lien': image_lien
      }

      data.append(obj)

    except:
      pass
  DF = pd.DataFrame(data)
  df = pd.concat([DF,df], axis = 0).reset_index(drop = True)
```
